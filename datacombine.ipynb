```python
import os
import pandas as pd
```

```python
filelist = os.listdir("./results/")
```

```python
df = []
for file in filelist:
    file_df = pd.read_csv("./results/"+file)
    file_df.drop(labels='Unnamed: 0',axis = 1,inplace=True)
    def f(x,filename):
        return(filename+'_'+x)
    file_df["model"] = file_df["model"].apply(f,args=(file.split('.')[0],))
    df.append(file_df)
```

```python
result = pd.concat(df)
```
